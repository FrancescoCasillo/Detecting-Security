```python
import json
import pandas as pd
import numpy as np
import os

```

```python
data = []
with open('.\Dataset\SecurityDiscoverer_classified_sentences.arff', "r") as f:
    for line in f:
        line = line.replace('\n', '')
        data.append(line.split(','))

securityAspects = ['Confidentiality', 'Integrity', 'Availability', 'Identification & Authentication', 'Accountability', 'Privacy']
    
df = pd.DataFrame(data)
df[12:]
```

```python
new_dataset = pd.DataFrame(index=np.arange(df[12:].shape[0]))

new_dataset["Requirement Text"] = None
new_dataset["Security Related"] = False

for aspect in securityAspects:
    new_dataset[aspect] = False

counter = 0
for row in df[12:].iterrows():
    new_dataset.loc[counter,"Requirement Text"] = row[1][7]
    if row[1][0] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[0]] = True

    elif row[1][1] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[1]] = True

    elif row[1][2] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[2]] = True

    elif row[1][3] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[3]] = True

    elif row[1][4] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[4]] = True

    elif row[1][5] == 'yes':
        new_dataset.loc[counter,"Security Related"] = True
        new_dataset.loc[counter,securityAspects[5]] = True
    counter += 1

new_dataset
```

```python
new_dataset.to_excel('.\Dataset\Requirements.xlsx',index=False)
```
