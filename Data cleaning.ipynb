## Cleaning Classified dataset

```python
import pandas as pd
import numpy as np
```

```python
data = []
with open('.\ClassifiedDataset\SecurityDiscoverer_classified_sentences.arff', "r") as f:
    for line in f:
        line = line.replace('\n', '')
        data.append(line.split(','))

securityAspects = ['Confidentiality', 'Integrity', 'Availability', 'Identification & Authentication', 'Accountability', 'Privacy']
    
df = pd.DataFrame(data)
df[12:]
```

```python
new_dataset_classified = pd.DataFrame(index=np.arange(df[12:].shape[0]))

new_dataset_classified["Requirement Text"] = None
new_dataset_classified["Security Related"] = False

for aspect in securityAspects:
    new_dataset_classified[aspect] = False

counter = 0
for row in df[12:].iterrows():
    new_dataset_classified.loc[counter,"Requirement Text"] = row[1][7]
    if row[1][0] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[0]] = True

    if row[1][1] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[1]] = True

    if row[1][2] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[2]] = True

    if row[1][3] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[3]] = True

    if row[1][4] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[4]] = True

    if row[1][5] == 'yes':
        new_dataset_classified.loc[counter,"Security Related"] = True
        new_dataset_classified.loc[counter,securityAspects[5]] = True
    counter += 1

new_dataset_classified
```

```python
new_dataset_classified.to_excel('.\ClassifiedDataset\RequirementsClassified.xlsx',index=False)
```

## Cleaning Original Datasets

```python
import json
import pandas as pd
import os
```

#### Example dataset opening

```python
data = json.load(open('OriginalDatasets\CCHIT Certified 2011 Ambulatory EHR Criteria 20110517_PUBLISH.json'))

df = pd.DataFrame(data["content"])
df
```

#### Opening all the dataset

```python
df = pd.DataFrame()

for file in os.listdir('./OriginalDatasets/'):
    print("Processing " + file + " ...")
    data = json.load(open("OriginalDatasets/"+file, encoding="utf8"))
    tmp = pd.DataFrame(data["content"])
    print("Processing "+str(tmp.shape[0])+" rows...")
    df = pd.concat([df, tmp])

df = df.reset_index(drop=True)
df
```

```python
secGoal = []

for row in df.iterrows():
    for obj in row[1]['securityObjectiveAnnotations']:
        if obj['securityObjective'] not in secGoal:
            secGoal.append(obj['securityObjective'])

print("Security goal: "+ str(secGoal))
print("Security aspects: "+ str(securityAspects))
```

```python
new_dataset_original = pd.DataFrame(index=np.arange(df.shape[0]))

new_dataset_original["Requirement Text"] = None
new_dataset_original["Security Related"] = False

for aspect in securityAspects:
    new_dataset_original[aspect] = False

counter = 0
for row in df.iterrows():
    new_dataset_original.loc[counter,"Requirement Text"] = row[1]['sentence']
    for obj in row[1]['securityObjectiveAnnotations']:
        if obj['securityObjective'] == 'CONFIDENTIALITY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[0]] = True
        if obj['securityObjective'] == 'ACCESS_CONTROL_IDENTITY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[3]] = True
        if obj['securityObjective'] == 'INTEGRITY' or obj['securityObjective'] == 'INTEGRITY_IMMUNITY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[1]] = True
        if obj['securityObjective'] == 'ACCOUNTABILITY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[4]] = True
        if obj['securityObjective'] == 'AVAILABILITY' or obj['securityObjective'] == 'AVAILABILITY_SURVIVABILITY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[2]] = True
        if obj['securityObjective'] == 'PRIVACY':
            new_dataset_original.loc[counter,"Security Related"] = True
            new_dataset_original.loc[counter,securityAspects[5]] = True
    counter += 1

new_dataset_original
```

```python
new_dataset_original.to_excel('.\DatasetClassified\RequirementsOriginal.xlsx',index=False)
```

```python
print("Security related requirements from Classified dataset: ")
print(len(new_dataset_classified[new_dataset_classified["Security Related"]==True]))
print("Not security related requirements from Classified dataset: ")
print(len(new_dataset_classified[new_dataset_classified["Security Related"]==False]))
print("------------------------------------------------------------")
print("Security related requirements from Original dataset: ")
print(len(new_dataset_original[new_dataset_original["Security Related"]==True]))
print("Not security related requirements from Original dataset: ")
print(len(new_dataset_original[new_dataset_original["Security Related"]==False]))
```
